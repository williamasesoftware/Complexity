# Comparison

```python
# Packages

import matplotlib.pyplot as plt
import time
import numpy as np
```

Functions definition

```python
def TA(n):
    return 1000 * n

# The big-O notation for this function is O(n), as the running time grows linearly with the input size n

def TB(n):
    return 200 * (n * np.log(n))

# The big-O notation for this function is O(n * log(n)), as the running time grows with the product of n and the logarithm of n.

def TC(n):
    return 10 * (n**2)

# The big-O notation for this function is O(n^2), as the running time grows with the square of the input size n.

def TD(n):
    return 2**n

# The big-O notation for this function is O(2^n), as the running time grows exponentially with the input size n.
```

Comparison with little size (input)

```python
x = np.linspace(1, 5, 100)

plt.plot(x, TA(x), label='TA(n)')
plt.plot(x, TB(x), label='TB(n)')
plt.plot(x, TC(x), label='TC(n)')
plt.plot(x, TD(x), label='TD(n)')

plt.xlabel('Input size n')
plt.ylabel('Running time')
plt.legend()
plt.show()
```

Comparison with big size (input)

```python
input_sizes = [100000,1000000,10000000,100000000,1000000000,10000000000]
complexities = [TA, TB, TC, TD]

for complexity in complexities:
    running_times = []
    for size in input_sizes:
        start = time.time()
        result = complexity(size)
        end = time.time()
        running_times.append(end - start)
    print("Running times for", complexity.__name__, ":", running_times)
    plt.plot(input_sizes, running_times, label=complexity.__name__)


plt.xlabel('Input Size')
plt.ylabel('Running Time (seconds)')
plt.legend()
plt.show()
```
